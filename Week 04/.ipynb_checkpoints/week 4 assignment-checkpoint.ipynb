Let's look at the properties of the network.

```python
stats = ox.basic_stats(gpg, circuity_dist='euclidean')
stats
```

In order to create isochrone maps, we will need use the centroid as a point to meausre the distance. By finding the minimum and maximum bounds for x and y, we could then calculate the centroid.

```python
minx, miny, maxx, maxy = gpg_nodes.geometry.total_bounds
print(minx)
print(miny)
print(maxx)
print(maxy)
```

```python
# calculate the centroid by minus min from max and subtract 2 to find the x and y coordinates
centroid_x = (maxx-minx)/2 + minx
centroid_y = (maxy-miny)/2 + miny
print(centroid_x)
print(centroid_y)
```

After finding centroid, we could find the nearest graph node.

```python
center_node = ox.distance.nearest_nodes(gpg,Y=centroid_y,X=centroid_x)
print('The id for the nearest node is ' + str(center_node))
```

let's print the center node.

```python
gpg_nodes.loc[[center_node]]
```

Now we can try to plot the map.

```python
fig, ax = plt.subplots(figsize=(10,10))

gpg_edges.plot(ax=ax,
               linewidth=0.5,
               edgecolor='gray', 
               zorder=10)

gpg_nodes.plot(ax=ax, 
               markersize=2, 
               zorder=20)

gpg_nodes.loc[[center_node]].plot(ax=ax,
                                  color='r', 
                                  zorder=30)

ax.axis('off')

# add a basemap
ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron,zoom=14)
```

Now we will graph isochromes by finding the distance from central node.

```python
# first is to calculate the time it takes to travel that edge
gpg_edges['walk_time'] = gpg_edges['length']/meters_per_minute
```

```python
## now we are going to assign colors for the isochrome maps.
iso_colors = ox.plot.get_colors(n=len(trip_times), 
                                cmap='Wistia', 
                                start=0, 
                                return_hex=True)
print(trip_times)
print(iso_colors)
```

```python
## reverse the colors, so darker colors match with longer time
trip_times.sort(reverse=True)
print(trip_times)
print(iso_colors)
```

```python
time_color = list(zip(trip_times, iso_colors))
time_color
```

```python
for time,color in list(time_color):
    print('The color for '+str(time)+' minutes is ' + color)
```

With this loop, we can calculate how many nodes are there within each one.

```python
for time, color in list(time_color):
    subgraph = nx.ego_graph(gpg, center_node, radius=time)

    print('There are ' + str(len(subgraph.nodes())) + ' nodes within ' + str(time) + ' minutes ')
    
    for node in subgraph.nodes():
        gpg_nodes.loc[node,'time'] = str(time) + ' mins'
        gpg_nodes.loc[node,'color'] = color
```

```python
gpg_nodes.sample(10)
```

```python
## Give 'NaN' a color
gpg_nodes['color'].fillna('#cccccc', inplace=True)
```

```python
gpg_nodes.plot(figsize=(10,10),
               color=gpg_nodes['color'],
               markersize=3)
```

Since we have the polygons, we don't need nodes anymore.

```python
isochrones = gpg_nodes.dissolve("time")
isochrones
```

```python
isochrones = isochrones.convex_hull.reset_index()
isochrones
```

```python
isochrones.columns=['time','geometry']
```
